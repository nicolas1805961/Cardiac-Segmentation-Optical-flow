```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import os

# Specify the path to your TensorBoard log directory
log_dir_1 = r"C:\Users\Portal\Documents\Isensee\nnUNet\nnunet\medis_model\Epoch_Train_vs_val_loss_Train\events.out.tfevents.1716588764.r3i4n6.521879.2"
log_dir_2 = r"C:\Users\Portal\Documents\Isensee\nnUNet\nnunet\medis_model\Epoch_Train_vs_val_loss_Val\events.out.tfevents.1716588764.r3i4n6.521879.3"

# Initialize the list to hold loss values and their corresponding steps
loss_values_1 = []
loss_values_2 = []
steps = []

# Load the TensorBoard log files
for event in tf.compat.v1.train.summary_iterator(log_dir_1):
    for value in event.summary.value:
        if value.tag == 'Epoch/Train_vs_val_loss':  # Replace 'loss' with your actual tag name if different
            loss_values_1.append(value.simple_value)
            steps.append(event.step)

# Load the TensorBoard log files
for event in tf.compat.v1.train.summary_iterator(log_dir_2):
    for value in event.summary.value:
        if value.tag == 'Epoch/Train_vs_val_loss':  # Replace 'loss' with your actual tag name if different
            loss_values_2.append(value.simple_value)

# Convert lists to numpy arrays for easier manipulation
loss_values_1 = np.array(loss_values_1)
loss_values_2 = np.array(loss_values_2)
steps = np.array(steps)

print(loss_values_1.shape)
print(loss_values_2.shape)
print(steps.shape)

# Plot the data using Matplotlib
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(steps, loss_values_1, label='Training Loss')
ax.plot(steps, loss_values_2, label='Validation Loss')
ax.set_xlabel('epoch')
ax.set_ylabel('Loss')
ax.set_title('Training and validation Loss Over Time')
ax.legend()
ax.grid(True)
plt.tight_layout()

plt.savefig(os.path.join(r'C:\Users\Portal\Documents\IRBM', 'loss_plot.tiff'), dpi=600, transparent=False)
plt.show()
```
